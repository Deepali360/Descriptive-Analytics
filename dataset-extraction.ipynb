# using append() inside a loop can be inefficient because it creates a new DataFrame each time. #To optimize it, store the Df in a list and concatenate them at the end using pd.concat()

```python
final_df = []  # Using a list to accumulate dataframes
counter = 1

for file in tqdm(filenames):
    with open(file, 'r') as f:
        #load the yaml file and normalise the JSON Content
        df=pd.json_normalize(safe_load(f))
        # Add Match_id column
        df['match_id'] = counter
        # Append Dataframe to the list
        final_df.append(df)
        counter +=1
        
# Concatenate all data frames in the list into a single data frame

final_df = pd.concat(final_df, ignore_index= True) 

# display final Dataframe

final_df
```

```python
import pandas as pd

# Initialize the main DataFrame to store ball-by-ball data for 1st innings of each match
count = 1  # Keeps track of match number or match_id
Main_df = pd.DataFrame()

# Loop through each row of final_df, which contains info about matches
# Each row in final_df represents data for one match
for index, row in final_df.iterrows():

    # Initialize empty lists to store ball-by-ball details for the current match
    ball_of_match = []  # List to store ball numbers
    batsman = []        # List to store batsman names
    bowler = []         # List to store bowler names
    runs = []           # List to store runs scored on each ball
    player_of_dismissed = []  # List to store the player dismissed on each ball (if any)
    teams = []          # List to store team names
    batting_team = []   # List to store the batting team
    match_id = []       # List to store the match id
    city = []           # List to store the city where the match was played
    venue = []          # List to store the venue name

    # Iterate over each delivery in the 1st innings
    for ball in row['innings'][0]['1st innings']['deliveries']:
        # For each ball in the delivery, extract details
        for key in ball.keys():  # ball keys represent the ball number, e.g., '0.1', '0.2'
            match_id.append(count)  # Append the match id (current match number)
            batting_team.append(row['innings'][0]['1st innings']['team'])  # Append the batting team
            teams.append(row['info.teams'])  # Append both teams playing in the match
            ball_of_match.append(key)  # Append the ball number (e.g., '0.1', '0.2')
            batsman.append(ball[key]['batsman'])  # Append batsman name for this ball
            bowler.append(ball[key]['bowler'])  # Append bowler name for this ball
            runs.append(ball[key]['runs']['total'])  # Append runs scored on this ball
            city.append(row['info.city'])  # Append the city where the match was played
            venue.append(row['info.venue'])  # Append the venue where the match was played

            try:
                # Try to get the player dismissed on this ball, if any
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except:
                # If no player was dismissed, append '0' to the list
                player_of_dismissed.append('0')

    # Create a temporary DataFrame (loop_df) for the current match's ball-by-ball data
    loop_df = pd.DataFrame({
        'match_id': match_id,
        'teams': teams,
        'batting_team': batting_team,
        'ball': ball_of_match,
        'batsman': batsman,
        'bowler': bowler,
        'runs': runs,
        'player_dismissed': player_of_dismissed,
        'city': city,
        'venue': venue
    })

    # Instead of append(), use pandas.concat to combine the current loop_df with Main_df
    Main_df = pd.concat([Main_df, loop_df], ignore_index=True)

    # Increment the match count for the next match
    count += 1

# The final Main_df now contains ball-by-ball data for all the matches in final_df's first innings.
```

```python
def bowl(row):
    for team in row['teams']:
        if team != row['batting_team']:
            return team
```

```python
Main_df['bowling_team'] = Main_df.apply(bowl,axis=1)
```
